```python
a = pd.read_excel('rel_very_new.xlsx', index_col=0)
```

```python
result = pd.DataFrame()
b = a.groupby(['CALLING','CALLED'])['DURATION'].count().reset_index()
unique_mob = b.CALLING.unique().tolist()
top3_freq_mob = pd.DataFrame()
for i in unique_mob:
    temp = b[b.CALLING == i].copy()
    temp = temp.sort_values(by=['DURATION'], ascending=False)
    temp = temp.iloc[:3,:].reset_index(drop=True)
    result = pd.concat([result, temp], ignore_index=True)
    
```

```python
result['CALLING'] = crypt(result.CALLING.tolist(), 'dc')
result['CALLED'] = crypt(result.CALLED.tolist(), 'dc')
```

```python
c1 = a. groupby(['CALLING','VILLAGE_NAME'])['DURATION'].count().reset_index().sort_values(by='DURATION', ascending=False).drop_duplicates(subset=['CALLING'])
c2 = a. groupby(['CALLING','DISTRICT_NAME'])['DURATION'].count().reset_index().sort_values(by='DURATION', ascending=False).drop_duplicates(subset=['CALLING'])
c3 = a. groupby(['CALLING','PROVINCE_NAME'])['DURATION'].count().reset_index().sort_values(by='DURATION', ascending=False).drop_duplicates(subset=['CALLING'])
```

```python
re = c1.merge(c2, 'left', on='CALLING')
re = re.merge(c3, 'left', on='CALLING')
```

```python
re.drop(columns=['DURATION_x','DURATION_y','DURATION'], inplace=True)
```

```python
result.to_excel('frequent_call.xlsx')
re.to_excel('location.xlsx')
```
